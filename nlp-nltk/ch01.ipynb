```python
def analyze_distribution(text):
    fd1 = FreqDist(text)
    print(fd1)
    print(fd1.most_common(20))
    print(fd1.hapaxes()[:10])
    fd1.plot(25, cumulative=True)

analyze_distribution(text1)
```

```python
def long_words(text):
    vocab = set(text)
    long_words = [ word for word in vocab if len(word) > 15 ]
    if len(long_words) == 0:
        print("no long words found")
        return
    print(sorted(long_words))
    long_ness_nouns = [ word for word in long_words if word.endswith("ness") ]
    print("long ness_nouns:")
    print(long_ness_nouns)
    # there is a bug here
    # words like finess
    ness_nouns = [ w for w in vocab if w.endswith("ness") ]
    print("All ness_nouns count: ", len(ness_nouns))
    # dispersion plot
    text.dispersion_plot(long_words)

long_words(text1)
```

```python
def l_long_f_frequent_words(text, l, f):
    fdist = FreqDist(text)
    found = [w for w in set(text) if len(w) > l and fdist[w] > f]
    print(sorted(found))

l_long_f_frequent_words(text5, 7, 7)
```

```python
# 3.4 Counting other things
def length_freq_dist(t):
    # length frequency distribution of text
    lens = [len(w) for w in t]
    lfd = FreqDist(lens)
    print(lfd)
    lfd.plot(25, cumulative=True)
    print(lfd.most_common(10))
    lfd.tabulate(5)

length_freq_dist(text1)
```

```python
length_freq_dist(text2)
```

```python
length_freq_dist(text5)
```
